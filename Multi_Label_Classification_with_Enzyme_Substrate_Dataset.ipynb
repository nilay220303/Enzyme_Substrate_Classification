```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Loading Training Dataset**

```python
df = pd.read_csv("enzyme_train.csv")
df.head()
```

```python
train_df = df.drop("id", axis=1)
```

```python
label_names = list(df.columns[-6:])
label_names
```

# **Preparing Training Tensor Frame**

---



```python
!pip install tensorflow_decision_forests
```

```python
import tensorflow_decision_forests as tfdf
import tensorflow as tf
```

```python
features = dict(train_df.drop(label_names, axis=1))
labels   = dict(train_df[label_names])
```

```python
train_tf = tf.data.Dataset.from_tensor_slices((features, labels)).batch(100)
```

# **Building Model**

---



```python
from tensorflow_decision_forests.keras import GradientBoostedTreesModel
```

```python
gbtm_model = GradientBoostedTreesModel(multitask = [tfdf.keras.MultiTaskItem(label = i, task = tfdf.keras.Task.CLASSIFICATION)
for i in label_names], verbose = 1,)
```

```python
gbtm_model.fit(train_tf)
```

```python
gbtm_model.summary()
```

# **Preparing Testing TensorFrame**

---



```python
test_df = pd.read_csv("enzyme_test.csv")
test_df.head()
```

```python
test_feats = dict(test_df.drop("id", axis=1))
```

```python
test_tf = tf.data.Dataset.from_tensor_slices((test_feats, {})).batch(100)
```

# **Test Predictions**

---



```python
preds = gbtm_model.predict(test_tf)
```

```python
preds
```

# **Submission File**

---



```python
submission = pd.DataFrame({"id": test_df["id"], "EC1": preds["EC1"].flatten(), "EC2": preds["EC2"].flatten()})
submission.head()
```

```python
submission.to_csv("submission.csv", index = False)
```
